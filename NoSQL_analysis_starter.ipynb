### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define a simpler query
query_basic = {'RatingValue': 5}

# Execute the query and fetch all documents
basic_results = list(establishments.find(query_basic))

# Convert the list of documents to a Pandas DataFrame
basic_results_df = pd.DataFrame(basic_results)

# Display the DataFrame
basic_results_df.head()
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define the query and sort criteria
degree_search = 0.01
latitude = 51.490142
longitude = 0.08384
query3 = {
    'RatingValue': 5,
    'geocode.latitude': {'$lte': latitude + degree_search, '$gte': latitude - degree_search},
    'geocode.longitude': {'$lte': longitude + degree_search, '$gte': longitude - degree_search}
}
sort_criteria = [('scores.Hygiene', 1)]

# Execute the query and sort, then fetch all documents
top_5_nearby_establishments = list(establishments.find(query3).sort(sort_criteria).limit(5))

# Convert the list of documents to a Pandas DataFrame
top_5_nearby_df = pd.DataFrame(top_5_nearby_establishments)

# Display the DataFrame
top_5_nearby_df
```

```python
# Convert result to Pandas DataFrame
pd.DataFrame(establishments.find(query3).sort(sort_criteria).limit(5))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline_variation = [
    {'$match': {'scores.Hygiene': 0}},
    {'$group': {'_id': '$LocalAuthorityName', 'total': {'$sum': 1}}},
    {'$sort': {'total': -1}}
]

result_variation = list(establishments.aggregate(pipeline_variation))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(result_variation)}")

# Print the first 10 results
pprint(result_variation[:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_zero_df = pd.DataFrame(result_variation)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hygiene_zero_df)}")

# Display the first 10 rows of the DataFrame
print(hygiene_zero_df.head(10))
```
